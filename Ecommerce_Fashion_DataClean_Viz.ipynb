### Ecommerce Fashion

##### Importing Library and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
mens = pd.read_csv('/content/Asosmenfashion.csv')
womens = pd.read_csv('/content/AsosWomenfashion.csv')
```

#### Data Cleaning and Transformation

##### Mens

```python
mens.head(3)
```

```python
mens.rename(columns={'rrp':'recommended_price'},inplace=True)
mens.shape
```

```python
mens.isna().sum()
```

```python
mens['product_id']=mens['product_id'].astype('object')
mens['colour']=mens['colour'].astype('object')
mens['productCode']=mens['productCode'].astype('object')

mens['colour']=mens['title'].str.split().str[-1]

#mens['previous_price'].fillna(0,inplace=True)
mens.fillna({'previous_price':0},inplace=True)
mens['previous_price']=mens['previous_price'].astype('float')
#mens['recommended_price'].fillna(0,inplace=True)
mens.fillna({'recommended_price':0},inplace=True)
mens['recommended_price']=mens['recommended_price'].astype('float')

mens['discount_%']=mens.apply(lambda x: (x['previous_price']-x['current_price'])/x['previous_price']*100 if x['previous_price']!=0 else 0,axis=1)
mens['discount_%']=mens['discount_%'].astype('float')
categorical_feature = [i for i in mens.columns if mens[i].dtypes == 'O']
numerical_feature = [i for i in mens.columns if mens[i].dtypes != 'O']
```

```python
mens.describe()
```

```python
categorical_feature,numerical_feature
```

##### Womens

```python
womens.head(3)
```

```python
womens.rename(columns={'rrp':'recommended_price'},inplace=True)
womens.shape
```

```python
womens.isna().sum()
```

```python
womens['product_id']=womens['product_id'].astype('object')
womens['colour']=womens['colour'].astype('object')
womens['productCode']=womens['productCode'].astype('object')

womens['colour']=womens['title'].str.split().str[-1]

#womens['previous_price'].fillna(0,inplace=True)
womens.fillna({'previous_price':0},inplace=True)
womens['previous_price']=womens['previous_price'].astype('float')
#womens['recommended_price'].fillna(0,inplace=True)
womens.fillna({'recommended_price':0},inplace=True)
womens['recommended_price']=womens['recommended_price'].astype('float')

womens['discount_%']=womens.apply(lambda x: (x['previous_price']-x['current_price'])/x['previous_price']*100 if x['previous_price']!=0 else 0,axis=1)
womens['discount_%']=womens['discount_%'].astype('float')
categorical_feature = [i for i in womens.columns if womens[i].dtypes == 'O']
numerical_feature = [i for i in womens.columns if womens[i].dtypes != 'O']
```

#### Answering Qustions using Vizualization

```python
mens_brand = mens['brand_name'].value_counts()
womens_brand = womens['brand_name'].value_counts()
```

```python
mens_brand.head(3)
```

```python
womens_brand.head(3)
```

```python
men_avg_curr_price_brand = mens.groupby('brand_name')['current_price'].mean()
women_avg_curr_price_brand = womens.groupby('brand_name')['current_price'].mean()
```

```python
men_avg_curr_price_brand.head(3)
```

```python
women_avg_curr_price_brand.head(3)
```

Q1. Which comapny has the highest and the lowest Avg. pricing for men and women?

```python
px.bar(men_avg_curr_price_brand,color=men_avg_curr_price_brand.index,title='Avg. Price By Brands (Mens)',hover_name=men_avg_curr_price_brand.index)
```

```python
px.bar(women_avg_curr_price_brand,color=women_avg_curr_price_brand.index,title='Avg. Price By Brands (Womens)',hover_name=women_avg_curr_price_brand.index)
```

Q2. What is the total share of inventory on offer on website for men and women?



```python
px.pie(mens_brand,values=mens_brand.values,names=mens_brand.index,color=mens_brand.index,title='Market Share (Mens)')
```

```python
px.pie(womens_brand,values=womens_brand.values,names=womens_brand.index,color=womens_brand.index,title='Market Share (Womens)')
```

```python
fashion_data = pd.concat([mens,womens],keys=['Mens','Womens'])
fashion_data.reset_index(inplace=True)
fashion_data.drop(columns='level_1',inplace=True)
fashion_data.rename(columns={'level_0':'category'},inplace=True)
```

```python
fashion_data.head(3)
```

Q3. Which gender has more products on offer through the website?


```python
count_products_category = fashion_data.groupby('category')['current_price'].count()
```

```python
px.bar(count_products_category,color=count_products_category.index,title='Avialability of Products By Category',hover_name=count_products_category.index)
```

Q4. Which gender has cheaper clothes available on Website?

```python
avg_products_category = fashion_data.groupby('category')['current_price'].mean()
```

```python
px.bar(avg_products_category,color=avg_products_category.index,title='Avg. Pricing of Products By Category',hover_name=avg_products_category.index)
```

Q5. Which Brands have more discount on the website?

```python
discount_brand = fashion_data.groupby('brand_name')['discount_%'].mean()
```

```python
px.bar(discount_brand,color=discount_brand.index,title='Avg. Discount of Products By Brand',hover_name=discount_brand.index)
```
